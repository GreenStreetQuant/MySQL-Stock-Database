```
import pandas as pd 
from pandas import DataFrame
import sys
import numpy as np 
import pandas_datareader.data as pdr 
import yfinance as yf 
from google.colab import files


data = files.upload()
```

```

#Reading a csv file from nasdaq with all tradable securities
df = pd.read_csv('nasdaq (6).csv')
columns = ['Security Name', 'Market Category', 'Test Issue', 'Financial Status', 'Round Lot Size', 'ETF', 'NextShares']
ticker_csv = df.drop(columns=columns)
#print(ticker_csv.head())

#Creating a list for the tickers to fetch stock data
def csv_to_list(csv):

	ticker_list = []
	
	for index, rows in ticker_csv.iterrows():
		#Create list for the current row
		my_list = [rows.Symbol]

		#Append the list to the final list
		ticker_list.append(my_list)

	return ticker_list

data = csv_to_list(ticker_csv)

output_list = []

#Removing nested lists 
def reemovNestings(csv_list):

    for i in csv_list:
        if type(i) == list:
            reemovNestings(i)
        else:
            output_list.append(i)

reemovNestings(data)

clean_data = output_list

#Function to get stock data
def get_data(tickers):

    for symbol in tickers:
        try:
            tickers = yf.download(tickers, start='2019-01-01', end='2019-01-31', groupby='ticker')
            return tickers 
        except:
            print('Stock was not found'+symbol)
            pass



stock_data = get_data(clean_data)
```

```
df = pd.DataFrame(data=stock_data)
print(df.info()) 
```

```
unstack_df = pd.DataFrame(df.unstack())
unstack_df = unstack_df.reset_index()
```

```
df_name = unstack_df.rename(columns = {'level_0': 'Metrics', 'level_1': 'Ticker',
                                       0: 'Values'})
print(df_name)
```

```
df_name.to_csv(r'stock_01_19_02.csv')
```
